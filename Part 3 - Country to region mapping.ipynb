```python
regions = pd.read_csv('countries regions2.csv')
```

### Convert the regions dataframe into a dictionary

This will will be used to map countries to their regions/continent

```python
regions = regions.set_index('name')
```

```python
regions.head(2)
```

```python
#The dictionary
list(regions.to_dict().values())[0]
```

### Create new column 'region' by mapping country to its region using the mapping

```python
regional_mapping = list(regions.to_dict().values())[0]
```

```python
data['region'] = data['Country'].map(regional_mapping)
```

```python
data[['Country', 'region']]
```

There was only one row where 'region' was null and 'country' was not null. There is no country called Nomadic

```python
data[data.loc[:, 'region'].isnull()][['Country', 'region']]['Country'].unique()
```
